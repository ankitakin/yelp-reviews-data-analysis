```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

checkin_df=pd.read_json("/Users/jieyizhang/Downloads/yelp_dataset/yelp_academic_dataset_checkin.json", lines=True)
```

```python
busCheck_df = business_df.merge(checkin_df,left_on='business_id',right_on='business_id',how='left')
print(len(busCheck_df))
busCheck_df
```

```python
us_df = busCheck_df[(busCheck_df['state']=='AZ')|(busCheck_df['state']=='NV')|(busCheck_df['state']=='NC')|(busCheck_df['state']=='OH')|\
            (busCheck_df['state']=='PA')|(busCheck_df['state']=='WI')|(busCheck_df['state']=='IL')\
           |(busCheck_df['state']=='SC')]
```

```python
weekdays = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
hours = []
keys = []
buckets = [0] * 168
for x in range(24):
    hours.append(str(x))

for w in weekdays:
    for h in hours:
        keys.append(w+'-'+h)

# print(keys[1])
# print(buckets)

for i in range(168):
    key = keys[i]
    for j in us_df.index:
        occ_dict = us_df.loc[j,'time']
        try:
            val = occ_dict[key]
            buckets[i] = buckets[i] + val
        except:
            pass
        
print(buckets)
```

```python
# MonKey = keys[0:24]
# TueKey = keys[24:48]
# WedKey = keys[48:72]
# ThuKey = keys[72:96]
# FriKey = keys[96:120]
# SatKey = keys[120:144]
# SunKey = keys[144:168]

plt.figure(1)
plt.rcParams['figure.figsize']=(20,20)
plt.xlabel('Time')
plt.ylabel('Number of Checkins')
plt.subplot(421)
plt.plot(buckets[0:24])
plt.xticks(np.arange(0, 24, 1))

# plt.xlabel('Countries')
# plt.ylabel('Population in million')
plt.title('Checkin on Monday (11 Top States in the U.S.)')
plt.subplot(422)
plt.plot(buckets[24:48])
plt.xticks(np.arange(0, 24, 1))
plt.title('Checkin on Tuesday (11 Top States in the U.S.)')

plt.subplot(423)
plt.plot(buckets[48:72])
plt.xticks(np.arange(0, 24, 1))

# plt.xlabel('Countries')
# plt.ylabel('Population in million')
plt.title('Checkin on Wednesday (11 Top States in the U.S.)')
plt.subplot(424)
plt.plot(buckets[72:96])
plt.xticks(np.arange(0, 24, 1))
plt.title('Checkin on Thursday (11 Top States in the U.S.)')

plt.subplot(425)
plt.plot(buckets[96:120])
plt.xticks(np.arange(0, 24, 1))

# plt.xlabel('Countries')
# plt.ylabel('Population in million')
plt.title('Checkin on Friday (11 Top States in the U.S.)')
plt.subplot(426)
plt.plot(buckets[120:144])
plt.xticks(np.arange(0, 24, 1))
plt.title('Checkin on Saturday (11 Top States in the U.S.)')
plt.subplot(428)
plt.plot(buckets[144:168])
plt.xticks(np.arange(0, 24, 1))
plt.title('Checkin on Sunday (11 Top States in the U.S.)')


plt.savefig('/Users/jieyizhang/Downloads/yelp_dataset/Checkin American Top 11 Sep.png')
```

```python
t = np.arange(0, 24, 1)
plt.plot(t,buckets[0:24] , 'y', t, buckets[24:48], 'y', t,buckets[48:72] , 'y',\
        t,buckets[72:96] , 'y', t, buckets[96:120], 'orange', t,buckets[120:144] , 'r',\
        t,buckets[144:168],'b')
plt.rcParams['figure.figsize']=(11,6)
plt.xlabel('Time')
plt.ylabel('Number of Checkins')
plt.xticks(np.arange(0, 24, 1))
plt.title('Checkin in Different Weekdays (11 Top States in the U.S.)')
xposition = [1, 19]
for xc in xposition:
    plt.axvline(x=xc, color='k', linestyle='--')
plt.savefig('/Users/jieyizhang/Downloads/yelp_dataset/Checkin American Top 11 Tog.png')
```
