```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

with open('/Users/jieyizhang/Downloads/yelp_dataset/yelp_academic_dataset_review.json') as json_file:      
    data = json_file.readlines()
    data = list(map(json.loads, data)) 

review_df = pd.DataFrame(data)
review_df
```

```python
reviewnum_user = review_df.groupby('user_id')['review_id'].count().to_frame()
reviewnum_user
```

```python
reviewnum_user[reviewnum_user['review_id']==reviewnum_user['review_id'].max()]
```

```python
user_dictc = {}
for user in review_df.user_id:
    if(user in user_dictc):
        user_dictc[user] += 1
    else:
        user_dictc[user] = 1
        

s5 = [(k, user_dictc[k]) for k in sorted(user_dictc, key=user_dictc.get, reverse=True)]


for k, v in s5:   
    print(k, v)
```
